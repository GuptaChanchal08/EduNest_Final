```python
df['Direction'].unique()
```

```python
def classify_position(x):
    x = x.upper()
    if 'LONG' in x:
        return 'LONG'
    elif 'SHORT' in x:
        return 'SHORT'
    else:
        return 'OTHER'

df['Position'] = df['Direction'].apply(classify_position)
```

```python
df['Position'].value_counts(normalize=True) * 100
```

```python
longs = (df['Position'] == 'LONG').sum()
shorts = (df['Position'] == 'SHORT').sum()
ratio=longs/shorts
ratio
```
